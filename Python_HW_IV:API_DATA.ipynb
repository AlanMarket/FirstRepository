```
## I found this code block on gitHub and it was written by Freddie Zhang
## and modified to work with how I began the code.
HH_Income = []
Pop_by_age = []
Pop_by_Hispanic = []
for i, row in data.iterrows():
    Income = parsedjson['data'][data['geoid'][i]]['B19001']['estimate']['B19001011']# B19001011 $50,000 to $59,999
    + parsedjson['data'][data['geoid'][i]]['B19001']['estimate']['B19001012'] # B19001012 $60,000 to $74,999
    HH_Income.append(Income)
    Pop = parsedjson['data'][data['geoid'][i]]['B01001']['estimate']['B01001007']# B01001007 18 and 19 Years
    + parsedjson['data'][data['geoid'][i]]['B01001']['estimate']['B01001008']# B01001008 20 Years
    + parsedjson['data'][data['geoid'][i]]['B01001']['estimate']['B01001009']# B01001009 21 Years
    + parsedjson['data'][data['geoid'][i]]['B01001']['estimate']['B01001010']# B01001010 22 to 24 Years
    + parsedjson['data'][data['geoid'][i]]['B01001']['estimate']['B01001011']# B01001011 25 to 29 Years
    + parsedjson['data'][data['geoid'][i]]['B01001']['estimate']['B01001012']# B01001012 30 to 34 Years
    + parsedjson['data'][data['geoid'][i]]['B01001']['estimate']['B01001031']# B01001031 18 and 19 Years
    + parsedjson['data'][data['geoid'][i]]['B01001']['estimate']['B01001032']# B01001032 20 Years
    + parsedjson['data'][data['geoid'][i]]['B01001']['estimate']['B01001033']# B01001033 21 Years
    + parsedjson['data'][data['geoid'][i]]['B01001']['estimate']['B01001034']# B01001034 22 to 24 Years
    + parsedjson['data'][data['geoid'][i]]['B01001']['estimate']['B01001035']# B01001035 25 to 29 Years
    + parsedjson['data'][data['geoid'][i]]['B01001']['estimate']['B01001036']# B01001036 30 to 34 Years
    Pop_by_age.append(Pop)
    Hispanic = parsedjson['data'][data['geoid'][i]]['B03002']['estimate']['B03002012']# B03002012 Hispanic or Latino
    Pop_by_Hispanic.append(Hispanic)

data['Pop_by_Income'] = HH_Income
data['Pop_by_Age'] = Pop_by_age
data['Pop_by_Hispanic'] = Pop_by_Hispanic


```

```
print(data.loc[data['Pop_by_Income'] == max(data['Pop_by_Income'])]['NAME'])
print(data.loc[data['Pop_by_Age'] == max(data['Pop_by_Age'])]['NAME'])
print(data.loc[data['Pop_by_Hispanic'] == max(data['Pop_by_Hispanic'])]['NAME'])
```

```
incomePercent = []

for i in data['Pop_by_Income']:
  incomePercent.append(i/max(data['Pop_by_Income']))

data['IncomePercent'] = incomePercent
```

```
hispanPercent = []

for i in data['Pop_by_Hispanic']:
  hispanPercent.append(i/max(data['Pop_by_Hispanic']))

data['hispanPercent'] = incomePercent
```

```
agePercent = []

for i in data['Pop_by_Age']:
  agePercent.append(i/max(data['Pop_by_Age']))

data['agePercent'] = incomePercent
```

```
overallMean = []

for i, r in data.iterrows():
  overallMean.append((data['agePercent'][i]+data['hispanPercent'][i]+data['IncomePercent'][i])/3)
data['Overall Mean']= overallMean
```

```
print(data.loc[data['Overall Mean'] == max(data['Overall Mean'])]['NAME'])
```
